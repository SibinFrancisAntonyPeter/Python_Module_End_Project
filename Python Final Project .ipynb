## 1. Determine the distribution of Players across each team and calculate the percentage split relative to the total number of Players.

```python
df1
```

```python
# To get the team count

team_count = df1.groupby('Team')
team_count
```

```python
team_count.size()
```

#### Finding the percentage split

```python
t_count = df1['Team'].value_counts() # Count of players in each team
total_count = len(df1)  # Total number of players 
percentage = round((t_count/total_count)*100,2) # Calculating the percentage
print(percentage.head()) # Showing first 5 rows 
```

```python
count_df = pd.DataFrame({
    "Team" : t_count.index,
    "Players Count" : total_count,
    "Percentage" : percentage})
```

```python
count_df
```

```python
count_df.head()
```

#### Graphical Representation

```python
plt.figure(figsize = (9,7))
plt.bar(count_df["Team"],count_df["Percentage"],color= 'indigo',edgecolor='yellow')
plt.title ("Percentage split against each team")
plt.xlabel("Team Names")
plt.ylabel("Percentage")
plt.yticks([1,2,3,4,5,])
plt.xticks(rotation=90)
for i, v in enumerate(count_df['Percentage']):
    plt.text(i, v + 0.5, str(v), ha='center',rotation=90,fontsize=7)
plt.show()
```

### Inference : "New Orleans Pelicans" has the highest percentage of players.

#### Creating age groups

```python
bins = [10,15,20,25,30,35,40,45,50]
labels = ['10-15','15-20','20-25','25-30','30-35','35-40','40-45','45-50']
df1['Age_Group'] = pd.cut(df1['Age'], bins=bins, labels=labels, right=False) # Apply pd.cut() to create the 'Age_Group' column
Age_group = df1[['Name', 'Age', 'Age_Group']].copy() # Created a new DataFrame with Age_Group
Age_group
```

```python
# finding the predominant age group
Age_group.groupby('Age_Group').size()
```

#### Graphical Representation

```python
plt.figure(figsize = (9,7))
sns.histplot(Age_group['Age_Group'],bins = 9,color = 'steelblue', edgecolor = 'navy')
plt.title(" Distribution of age group among Employees")
plt.xlabel('Age Groups')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

### Inference : "25-30" is the predominant age group among the employees

## 4. Discover which team and position have the highest salary expenditure.

```python
df1
```

#### Finding the Team wise mean of salaries to find which Team has highest salary expenditure.


```python
Team_Salary = df1.groupby('Team')['Salary'].mean()
Team_Salary
```

```python
df1['Salary'] = df1['Salary'].apply(int)
pd.options.display.float_format = '{:,.0f}'.format
```

```python
# Sorting in ascending order
Team_Salary_sorted = Team_Salary.sort_values(ascending = False)
Team_Salary_sorted
```

#### Graphical Representation

```python
plt.figure(figsize = (9,7))
plt.bar(Team_Salary.index,Team_Salary.values,color = 'darkorange',edgecolor = 'greenyellow')
plt.title('Average Salary by Team')
plt.xlabel('Team Names')
plt.ylabel('Average salary (in Ten lakhs)')
plt.xticks(rotation=90)
plt.show()
```

### Inference : Team "Cleveland Cavaliers" have the highest salary.
